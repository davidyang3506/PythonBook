## 5.1 Introduction to pandas
`pandas` is a powerful and popular library in Python that is used for data manipulation and analysis. It provides data structures and functions needed to work with structured data seamlessly. The two primary data structures in pandas are Series and DataFrame .

#### Key Features of pandas:

1. Data Structures:
    - **Series**: A one-dimensional labeled array that can hold any data type (like integers, floats, strings, etc.). It's similar to a column in a spreadsheet or a database.
    
    - **DataFrame**: A two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It's like a table in a database or an Excel spreadsheet.


2. Data Manipulation:
    - `pandas` provides easy-to-use methods for filtering, grouping, pivoting, and transforming data. You can clean and prepare data for analysis with operations like merging, joining, and reshaping data.


3. Data Analysis:
    - You can perform descriptive statistics, handle missing data, and apply functions to your data efficiently.

    - It integrates well with other data analysis libraries like NumPy, SciPy, and Matplotlib, enabling complex analysis and visualizations.


4. Input and Output:
    - `pandas` supports reading and writing data from various file formats, including CSV, Excel, SQL databases, JSON, and more. This makes it easy to get data into and out of your data structures.


### 5.1.1 Creating DataFrames
We'll start by creating a DataFrame from a NumPy array.

```python
import numpy as np
import pandas as pd

# Creating a NumPy array
b = np.arange(20).reshape(4, 5)
print("NumPy array:\n", b)

# Creating a DataFrame from the NumPy array
df = pd.DataFrame(b)
print("DataFrame:\n", df)
print("Type of df:", type(df))
```

### 5.1.2 Vectorized Operations
A vectorized operation is an operation applied simultaneously to an entire array or sequence of data, rather than being executed in a loop that iterates over each element one at a time. In Python, particularly when using libraries like NumPy and pandas, vectorized operations are common and key to achieving fast and efficient computations.

Using vectorized operations often makes the code cleaner and more concise, as you can replace complex loops with a single line of code.

#### Scalar Multiplication:
Scalar multiplication is a basic mathematical operation where you multiply a single number (called a scalar) by each element in a vector or matrix. The scalar value scales or changes the magnitude of each element in the vector or matrix by the same factor.

```python
# Scalar multiplication
print("Scalar multiplication (3 * df):\n", 3 * df)
```

#### Element-wise Addition:
Element-wise addition is an operation where you add corresponding elements from two arrays, vectors, or matrices of the same size. The result is a new array, vector, or matrix where each element is the sum of the corresponding elements from the original inputs.

```python
# Element-wise addition
print("Element-wise addition (df + df):\n", df + df)
```

#### Element-wise Exponentiation:
Element-wise exponentiation is an operation where each element in an array, vector, or matrix is raised to the power of a corresponding element in another array, vector, or matrix, or to a single scalar power. The result is a new array, vector, or matrix where each element is the exponentiation of the original elements.

```python
# Element-wise exponentiation
print("Element-wise exponentiation (df ** 2):\n", df ** 2)
```

#### Other Operations:

```python
# Square root of each element
print("Square root of each element:\n", np.sqrt(df))
```

## 5.2 DataFrame Properties and Indexing
### 5.2.1 Columns and Index

You can access and modify the columns and index of a DataFrame.

```python
# Accessing columns and index
print("Columns:", df.columns)
print("Index:", df.index)
```

#### Code Breakdown:
- `df`: This is the variable that holds a DataFrame. A DataFrame is essentially a table-like data structure that contains rows and columns, similar to an Excel spreadsheet or a SQL table.

- `.columns`: This attribute returns an index object containing the column labels (names) of the DataFrame. It allows you to see or manipulate the names of the columns.

- `.index`: This attribute returns an object that represents the row labels of the DataFrame. By default, the index is a sequence of integers starting from 0, but it can also be a custom set of labels, such as dates, strings, or any other unique identifiers.

```python
# Renaming columns and index
df.columns = list('PQRST')
df.index = list('wxyz')
print("Renamed DataFrame:\n", df)
```

### 5.2.2 Accessing Data

You can access data in a DataFrame using various methods.

```python
# Accessing a single column by label
print("Column 'P':\n", df['P'])
```

```python
# Accessing a row by label
print("Row 'w':\n", df.loc['w'])
```

```python
# Accessing specific elements and slices
print("Element at 'w', 'P':\n", df.loc['w', 'P'])
print("Rows 'y' and 'z', columns 'Q' and 'R':\n", df.loc['y':'z', 'Q':'R'])
```

## 5.4 Visualization of Data
Visualization is a crucial part of data analysis. Pandas integrates with Matplotlib to provide easy plotting functionalities.
### 5.4.1 Creating a Time Series DataFrame
We'll use the Tiingo API to fetch financial data and create a DataFrame for visualization.

```python
from tiingo import TiingoClient
import os

# Set your Tiingo API key
os.environ['TIINGO_API_KEY'] = ''
client = TiingoClient()

# Fetching stock data for multiple tickers
tickers = ["IBM", "AMZN", "META"]
start_date = '2024-01-01'
end_date = '2024-02-01'
data = {}

for ticker in tickers:
    historical_prices = client.get_ticker_price(ticker, fmt='json', startDate=start_date, endDate=end_date, frequency='daily')
    dates = [entry['date'] for entry in historical_prices]
    prices = [entry['close'] for entry in historical_prices]
    data[ticker] = prices

data['Date'] = dates

# Converting the data to a pandas DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Ensure dates are in datetime format
df.set_index('Date', inplace=True)
print("DataFrame:\n", df.head())
```

### 5.4.2 Plotting Data

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

try:
    plt.style.use('ggplot')  
except OSError:
    print("ggplot style not found. Using default style.")
    plt.style.use('default')

get_ipython().run_line_magic('config', "InlineBackend.figure_format = 'svg'")

# Plotting the DataFrame
fig, ax = plt.subplots()
line_styles = ['-', '--', '-.', ':']
for i, ticker in enumerate(tickers):
    ax.plot(df.index, df[ticker], label=ticker, linestyle=line_styles[i % len(line_styles)])

ax.set_title("Stock Prices Over Time")
ax.set_xlabel("Date")
ax.set_ylabel("Prices")
ax.legend()

# Formatting date axis
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate()

plt.show()
```
